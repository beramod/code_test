```
You are given integers K, M and a non-empty array A consisting of N integers. Every element of the array is not greater than M.

You should divide this array into K blocks of consecutive elements. The size of the block is any integer between 0 and N. Every element of the array should belong to some block.

The sum of the block from X to Y equals A[X] + A[X + 1] + ... + A[Y]. The sum of empty block equals 0.

The large sum is the maximal sum of any block.

For example, you are given integers K = 3, M = 5 and array A such that:

  A[0] = 2
  A[1] = 1
  A[2] = 5
  A[3] = 1
  A[4] = 2
  A[5] = 2
  A[6] = 2
The array can be divided, for example, into the following blocks:

[2, 1, 5, 1, 2, 2, 2], [], [] with a large sum of 15;
[2], [1, 5, 1, 2], [2, 2] with a large sum of 9;
[2, 1, 5], [], [1, 2, 2, 2] with a large sum of 8;
[2, 1], [5, 1], [2, 2, 2] with a large sum of 6.
The goal is to minimize the large sum. In the above example, 6 is the minimal large sum.

Write a function:

def solution(K, M, A)

that, given integers K, M and a non-empty array A consisting of N integers, returns the minimal large sum.

For example, given K = 3, M = 5 and array A such that:

  A[0] = 2
  A[1] = 1
  A[2] = 5
  A[3] = 1
  A[4] = 2
  A[5] = 2
  A[6] = 2
the function should return 6, as explained above.

Write an efficient algorithm for the following assumptions:

N and K are integers within the range [1..100,000];
M is an integer within the range [0..10,000];
each element of array A is an integer within the range [0..M].
```

### 조건
- 정수 K, M과 N개의 정수로 구성된 비어있지 않은 배열 A
    - N, K: 1 ~ 100,000
    - M: 0 ~ 10,000
    - A: 0 ~ M
- 배열 A를 K개의 block으로 분할해야 한다.
- 분할시 block 합중 가장 큰 값이 가장 작은 경우가 되는 값을 리턴...
- block은 비어도 된다.

### 풀이
- minimal large sum의 lower bound와 upper bound를 찾는다.
- lower bound는 배열의 가장 큰값이다(block 하나당 원소 하나가 배정 되었을 때)
- upper bound는 A의 sum이다.(한 block에 모든 원소가 배정 되었을 때)
- lower bound를 left로, upper bound를 right로, 확인하고자 하는 large sum을 mid로 한다.
- mid(large sum)으로 K개만큼 분할이 가능한지 확인한다.(func valid)
- 최종 mid가 답

````python
def valid(max_sum, A, K):
    block_cnt = 1
    block_sum = 0
    
    for a in A:
        if block_sum + a <= max_sum:
            block_sum += a
        else:
            block_cnt += 1
            block_sum = a
    
    return (block_cnt <= K)
    
def solution(K, M, A):
    N = len(A)
    left = max(A)
    right = sum(A)
    
    if K == 1:
        return right
    if K >= N:
        return left

    answer = right
    
    while left <= right:
        mid = (left + right) // 2
        if valid(mid, A, K):
            right = mid - 1
            answer = mid
        else:
            left = mid + 1
    return answer
    
import sys
path = sys.path[0]
path = path[:path.index('code_test') + 9]
if path not in sys.path:
    sys.path.append(path)
from util.code_test_util import CodeTestUtil
code_test_util = CodeTestUtil()
code_test_util.set_multi_args(True)
code_test_util.solution = solution
code_test_util.add_data_case([3, 5, [2, 1, 5, 1, 2, 2, 2]], 6)
code_test_util.add_data_case([3, 3, [1, 2, 3]], 3)
code_test_util.add_data_case([1, 3, [1, 2, 3]], 6)
code_test_util.run()
````
