# 은 장식장을 구성하는 부분이며, . 은 비어있는 부분을 의미합니다.

장식장을 구성하는 #은 다음과 같은 특성이 있습니다.
1. 장식장 하나는 #이 두개이상 이어진 한 줄의 형태를 갖고 있습니다.
2. 장식장 하나를 구성하는 # 중에서 두 개의 #은 인전합 #의 개수가 한개 뿐입니다.
3. 2에서 설명한 두 개의 #을 제외한 다른 모든 #은 인전합 #의 개수가 두개입니다.

1  2  2  []  1
[] [] 2  []  2
2  2  2  []  2
2  [] []  1  2
2  2  1  []  []

위 그림은 각 #에 인접해 있는 다른 #의 개수를 표현한 그림입니다. 그림을 통해 환인할 수 있듯이, 인접한 #이 하나뿐인 장식장 격자는
